# A plot showing the representative amount of investment in each funding type

```python
#Impirt Libraries
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# groupby funding Type and average of the Amount
pie_frame = master_frame.groupby('funding_round_type')['raised_amount_usd'].mean()
## Check unique vaues in Funding Type
j= master_frame['funding_round_type'].unique()
j
```

```python
# groupby funding Type and average of the Amount
master_sum = master_frame.groupby('funding_round_type').agg( mean =('raised_amount_usd', lambda p: int(p.mean())))
master_sum.index
```

```python
master_sum.sort_values(by='mean',ascending =False)
```

```python
#Seaborn barplot for checking average in each 
sns.set_style("white")
bar,ax = plt.subplots(figsize=(25,35))
ax = sns.barplot(y=master_sum.index, x=master_sum['mean'], data=master_sum, ci=None, palette="muted",orient='h')
ax.set_title("Investment Average in each Funding Type", fontsize=50)
ax.set_ylabel ("Funding Type",fontsize = 40)
ax.set_xlabel ("Investment Avg",fontsize = 40)
for i, v in enumerate(master_sum['mean']):
    ax.text(v + 3, i + .25, str(v), 
            color = 'blue',fontsize=20)
ax.set_xticklabels(ax.get_xticklabels(), ha="right",fontsize=25)
ax.set_yticklabels(ax.get_yticklabels(), rotation=40, ha="right",fontsize=25)
```

```python
## PIE chart for inverstment in each funding Type
pie_frame = master_frame.groupby('funding_round_type')['raised_amount_usd'].mean()
pie_frame.plot.pie(subplots=True,figsize=(30, 15),title = 'Investment in each funding type',fontsize = 14)
```

```python
## % Investment in Top Funding Types
master_frame1 = master_frame.groupby('funding_round_type').mean().loc[['angel', 'seed', 'private_equity', 'venture']].sort_values(by = 'raised_amount_usd', ascending = False)
master_frame1
```

```python
master_frame1.plot.pie(subplots=True,figsize=(20,10),title = '% Investment in Top funding type',fontsize = 20,autopct="%.1f%%")
```
